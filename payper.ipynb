```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each unique value in the 'primary_site' column
# Keep only the top 22 most common cancer types
top_sites = combined_tcga_df["primary_site"].value_counts().nlargest(24).index
filtered_df = combined_tcga_df[combined_tcga_df["primary_site"].isin(top_sites)]
primary_site_counts = filtered_df['primary_site'].value_counts()

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=primary_site_counts.index, y=primary_site_counts.values, palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title("Distribution of Primary Sites")
plt.xlabel("Primary Site")
plt.ylabel("Count")
plt.tight_layout()



# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mapowanie skróconych nazw
class_short_labels = {
    "Hematopoietic and reticuloendothelial systems": "Hemato",
    "Bronchus and lung": "Lung",
    "Breast": "Breast",
    "Kidney": "Kidney",
    "Brain": "Brain",
    "Colon": "Colon",
    "Corpus uteri": "Corpus",
    "Skin": "Skin",
    "Prostate gland": "Prostate",
    "Stomach": "Stomach",
    "Bladder": "Bladder",
    "Liver and intrahepatic bile ducts": "Liver",
    "Pancreas": "Pancreas",
    "Ovary": "Ovary",
    "Uterus, NOS": "Uterus",
    "Cervix uteri": "Cervix",
    "Esophagus": "Esophagus",
    "Adrenal gland": "Adrenal",
    "Other and ill-defined sites": "Other",
    "Other and unspecified parts of tongue": "Tongue",
    "Connective, subcutaneous and other soft tissues": "Connective",
    "Larynx": "Larynx",
    "Rectum": "Rectum",
    'Thyroid gland':  'Thyroid gland',
    'Eye and adnexa': 'Eye',
    "Other and ill-defined sites in lip, oral cavity and pharynx": "Oral/Pharynx"
}

# Filtruj i zmapuj
top_sites = combined_tcga_df["primary_site"].value_counts().nlargest(24).index
filtered_df = combined_tcga_df[combined_tcga_df["primary_site"].isin(top_sites)].copy()
filtered_df["short_label"] = filtered_df["primary_site"].map(class_short_labels)

# Oblicz liczby
primary_site_counts = filtered_df["short_label"].value_counts()

# Wykres
plt.figure(figsize=(12, 6))
sns.barplot(x=primary_site_counts.index, y=primary_site_counts.values, palette="viridis")
plt.xticks(rotation=45, ha='right')
plt.title("Distribution of Primary Sites (Short Labels)")
plt.xlabel("Primary Site")
plt.ylabel("Count")
plt.tight_layout()

# Zapisz do PDF
plt.savefig("primary_site_distribution.pdf")

# Pokaż wykres
plt.show()
```
